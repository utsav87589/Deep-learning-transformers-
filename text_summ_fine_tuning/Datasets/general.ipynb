## Preparing the data

### 1. without_ctext

```python
import pandas as pd
import numpy as np
```

### train data

```python
df = pd.read_csv('news_summary.csv')
df.head()
```

```python
df.drop(['author', 'date', 'read_more', 'ctext'], axis = 1, inplace = True)
df.head()
```

```python
df.shape
```

```python
df_train = df.head(3000)
df_train
```

```python
df_train.to_csv('df_train.csv')
```

### eval data

```python
df_eval = df.iloc[3000 : 3801, :]
df_eval
```

```python
df_eval.to_csv('without_ctext/df_eval.csv')
```

### test data

```python
df_test = df.tail(713)
df_test
```

```python
df_test.to_csv('without_ctext/df_test.csv')
```
