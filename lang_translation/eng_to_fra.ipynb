## English to French

```python
import pandas as pd
import numpy as np
```

```python
data = 'words.txt'
english_words = []
```

```python
with open(data, 'r', encoding = 'UTF-8') as f : 
    lines = f.read().split('\n')

    for line in lines[:1000] : 
        eng_words, _ = line.split('\t')[:2]

        english_words.append(eng_words)
```

```python
english_words
```

### using the pre trained transformers

```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = 'Helsinki-NLP/opus-mt-en-fr'
```

```python
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
inputs = tokenizer(english_words, return_tensors = 'pt', padding = True, truncation = True)
```

```python
translated_tokens = model.generate(**inputs)
```

```python
french_texts = [tokenizer.decode(t, skip_special_tokens = True) for t in translated_tokens]
```

```python
french_texts
```

```python
for en, fr in zip(english_words, french_texts) : 
    print(f"{en} :: {fr}")
```
